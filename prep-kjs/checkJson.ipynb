```python
import json
import numpy as np
import pandas as pd
from collections import defaultdict
import re
```

```python
with open("/data1/lidc-idri/nodule_malignancy_scores.json", 'r') as f:
    annot = json.load(f)
```

```python
#group same roi in nodReader
#annot['LIDC-IDRI-0003']['slice_52']['reader_3'][0]['polygon']
#pid = 'LIDC-IDRI-0003'
nodReader = pd.DataFrame(columns=['pid','reader','noduleID','slice','xmin','xmax','ymin','ymax','score'])
# nodules[pid] = list()
for pid in annot.keys():
    pid = re.compile("(LIDC-IDRI-\d{4}).*").search(pid).group(1)
    for snum, val in sorted(annot[pid].items()):
        for reader, ann_list in val.items():
            # snuml[ann['noduleID']].append(snum)
            for ann in ann_list:
                # see LIDC-IDRI-0195 case, there's "09992" nodID case
                ann_noduleID = ann['noduleID']
                ann_noduleID = str(int(ann_noduleID)) if ann_noduleID.isdigit()  else ann_noduleID
                [(xmin,xmax),(ymin,ymax)] = list(map(lambda l: (min(l),max(l)),list(zip(*ann['polygon']))) )
                nodReader = pd.concat([nodReader,pd.DataFrame(
                    [{'pid':pid,'reader':reader,'noduleID':ann_noduleID,'slice':snum,
                    'xmin':xmin,'xmax':xmax,'ymin':ymin,'ymax':ymax,
                    'score':ann['score']}])],ignore_index=True)
            # print (ann['noduleID'],ann['score'])
            #print('polygon: ',ann['polygon'][:])
            # 
            # print('zipped: ',list(map(lambda l: (min(l),max(l)),list(zip(*ann['polygon']))) ) )
            # print (snum,len(val['reader_3']))


# print('snuml: ',snuml)
nodReader
```

```python
# nodG = nodReader.groupby('noduleID').apply(lambda g: list(g['slice'])) #lambda s: s.split('_')[1])
# nodG = nodG.apply(lambda l: list(map(lambda s: s.split('_')[1],l))) #.apply(pd.Series)
# nodG = nodG.apply(lambda l: [min(l),max(l)]).apply(pd.Series)
# nodG.columns = ['zidx_min','zidx_max']
nodG = nodReader.groupby(['pid','noduleID']).apply(
    lambda g: {'zcent':np.array(list(map(lambda s: s.split('_')[1],list(g['slice'])))).astype(int).mean(),
               'score':np.array(g['score']).mean()}).apply(pd.Series)
```

```python
nodules0 = pd.read_csv('list3.2.csv',dtype={"case": str})
nodules0['case'] = 'LIDC-IDRI-' + nodules0['case']
```

```python

#nodules0[nodules0.columns[list(nodules0.columns).index('nodIDs'):]]
# nodules0.columns.index('nodIDs')
iNidS = list(nodules0.columns).index('nodIDs')

realNodule = defaultdict(list)

for index, row in nodules0.iterrows():
    # if (row['case'] == 'LIDC-IDRI-0003'):
    # print(f"Index: {index}")
    # print(row[iNidS:].values)
    # [print(row.roi,nodG.loc[nid].to_list())  for nid in row[iNidS:] if (~pd.isna(nid)) ] #prepare to calculate average position
    # [print(row.roi, nodG.loc[nid])  for nid in row[iNidS:] if not pd.isna(nid) ] #prepare to calculate average position (z first)
    missNidInRoi=[]
    avg_roi = []
    for nid in row[iNidS:]:
        if not pd.isna(nid):
            if (row.case,nid) in nodG.index:
                # zcenter = np.array(nodG.loc[nid]).astype(int).mean()
                avg_roi.append(nodG.loc[(row.case,nid),['zcent','score']])
                # xmin,xmax = nodReader[nodReader.noduleID==nid][['xmin','xmax']].to_numpy()
            else:
                # missNidInRoi.append({'roi':row.roi, 'missed':nid})
                missNidInRoi.append(nid)
    if(len(avg_roi)==0): print ("empty roi: ",row.case,row.roi,row[iNidS:]) # can happen if entire row is not in nodreader (0132 has two records)
    # if(len(avg_roi)==0): print (row[iNidS:], nodG.loc[(row.case)])
    if(len(avg_roi)==0): continue
    zavg, mscore = pd.DataFrame(avg_roi).to_numpy().mean(axis=0)
    # zavg = np.array(z_roi).mean()
    # zavg = 
    zavgi = round(zavg)

    # print ("zavg, zavgi: ",zavg,zavgi)
    nodSliceAndRoi = nodReader[(nodReader.pid == row.case) &
                               (nodReader.noduleID.isin(row[iNidS:][lambda x: ~pd.isna(x)])) & 
                               (nodReader.slice==f"slice_{zavgi}")][['xmin','xmax','ymin','ymax']]
    if nodSliceAndRoi.empty:  print ("empty slice: ",row.case,row.roi, zavgi); continue  #slice output was skipped in kjs' pid 0509
    xmina, xmaxa, ymina, ymaxa = nodSliceAndRoi.to_numpy().mean(axis=0)
    
    realNodule['pid'].append(row.case)
    realNodule['roi'].append(row.roi)
    realNodule['zavgi'].append(zavgi)
    realNodule['xmina'].append(xmina)
    realNodule['xmaxa'].append(xmaxa)
    realNodule['ymina'].append(ymina)
    realNodule['ymaxa'].append(ymaxa)
    realNodule['mscore'].append(mscore)
    realNodule['malignancy'].append(1 if mscore >3 else 0)
    realNodule['missNidInRoi'].append(missNidInRoi)
    # zavg_roi = np.array(z_roi).mean()
        # xmin_roi = np.array(xmin).mean()
        # {'roi':row.roi, zavg_roi}            

#Why empty roi for 0132 ??
                

#combine nodReader to roi #nodules0 is dictionary
```
